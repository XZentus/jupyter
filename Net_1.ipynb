```python
import numpy as np
import matplotlib.pylab as plt

import random

numbers_raw = ['111101101101111',
               '001001001001001',
               '111001111100111',
               '111001111001111',
               '101101111001001',
               '111100111001111',
               '111100111101111',
               '111001001001001',
               '111101111101111',
               '111101111001111']

num5s = ['111100111000111',
         '111100010001111',
         '111100011001111',
         '110100111001111',
         '110100111001011',
         '111100101001111']


n5s = np.array([(list(map(lambda i: int(i), x))) for x in num5s])

weight = [0 for _ in range(15)]
nums = np.array([(list(map(lambda i: int(i), x))) for x in numbers_raw])
bias = 7

def routine(n):
    net = 0
    for i in range(15):
        net += n[i] * weight[i]
    return net >= bias

dec_counter = 0
inc_counter = 0

def decrease(n):
    global dec_counter
    dec_counter += 1
    for i in range(15):
        if n[i] == 1:
            weight[i] -= 1

def increase(n):
    global inc_counter
    inc_counter += 1
    for i in range(15):
        if n[i] == 1:
            weight[i] += 1

for i in range(10_000):
    option = random.randint(0, 9)
    if option != 5:
        if routine(nums[option]):
            decrease(nums[option])
    else:
        if not routine(nums[option]):
            increase(nums[option])

print(weight)
            
for i in range(len(nums)):
    print('{} == 5 : {}'.format(i, routine(nums[i])))

for n in n5s:
    print('Check 5? : {}'.format(routine(n)))

print(dec_counter, inc_counter)
```
